```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Ames = pd.read_csv('../ai-starter/bai09/Ames.csv')
# Tch d liu theo iu kin c hoc khng c iu ha trung tm
ac_prices = Ames[Ames['CentralAir'] == 'Y']['SalePrice']
no_ac_prices = Ames[Ames['CentralAir'] == 'N']['SalePrice']
# Thit lp biu 
plt.figure(figsize=(10, 6))
# V biu  histogram cho gi bn theo tnh trng iu ha
# V d liu "c iu ha" trc  hin th ng th t trong ch thch
plt.hist(ac_prices, bins=30, alpha=0.7, color='blue', edgecolor='blue', lw=0.5,
label='Gi bn - C iu ha')
mean_ac = np.mean(ac_prices)
plt.axvline(mean_ac, color='blue', linestyle='dashed', linewidth=1.5,
label=f'Trung bnh (C iu ha): ${mean_ac:.2f}')

plt.hist(no_ac_prices, bins=30, alpha=0.7, color='red', edgecolor='red', lw=0.5,
label='Gi bn - Khng iu ha')
mean_no_ac = np.mean(no_ac_prices)
plt.axvline(mean_no_ac, color='red', linestyle='dashed', linewidth=1.5,
label=f'Trung bnh (Khng iu ha): ${mean_no_ac:.2f}')
plt.title('Phn phi gi bn theo tnh trng iu ha trung tm',
fontsize=18)
plt.xlabel('Gi bn', fontsize=15)
plt.ylabel('S lng nh', fontsize=15)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
from scipy import stats
# T-test hai mu, khng gi nh phng sai bng nhau
t_stat, p_value = stats.ttest_ind(ac_prices, no_ac_prices, equal_var=False)
# Din gii kt qu
if p_value < 0.05:
    result = "Bc b H0  C khc bit gi bn"
else:
    result = "Khng bc b H0  Khng  bng chng v khc bit"
print(f"P-value = {p_value:.5f}  {result}")
```
